## atom
[Back to top](#top)

The atom module contains a set a relevant physical parameters. It can also compute some relevant properties, such as the - scalar - polarizability or the scattering rate (for a far off-resonant beam). So far, only the helium atom is implemented.

```python
# - example of properties
print('-'*30)
print('m = %.5g kg' % he.mass)  # atomic mass
print('g = %.1f' % he.lande_g_factor)  # g factor (for ground state)
print('a = %.5g m' % he.scattering_length)  # s-wave scattering length

# - example of calculations
# polarizability
wavelength = 1550e-9  # meter
alpha_SI =  he.get_alpha(wavelength, unit='SI')
alpha_au =  he.get_alpha(wavelength, unit='au')  # atomic units
print('-'*30)
print('polarizability at %i nm = %.2g SI' % (wavelength * 1e9, alpha_SI))
print('polarizability at %i nm = %.2g au' % (wavelength * 1e9, alpha_au))

# scattering rate
wavelength = 1550e-9  # meter
power = 6  # watts
waist = 135e-6  # meters
intensity = 2 * power / pi / waist ** 2
print('-'*30)
print('scattering rate = %.2g /s' % he.get_scattering_rate(intensity, wavelength))
```

## Crossed optical dipole trap - Lattice Helium

```python
# -- settings
# parameters
power_ODT1 = 18
waist_ODT1 = 133e-6
half_angle_ODT1 = 20 * pi / 180  # radians

power_ODT2 = 8
waist_ODT2 = 63e-6
half_angle_ODT2 = -20 * pi / 180  # radians
 

# beams
# NB : wavelength is 1550nm by default
beamODT1 = {'waist_value': waist_ODT1,
         'power': power_ODT1,
         'theta': pi / 2,  # i.e. in horizontal plane
         'phi': half_angle_ODT1,
         'label': 'ODT1'}

beamODT2 = {'waist_value': waist_ODT2,
         'power': power_ODT2,
         'theta': pi / 2,  # i.e. in horizontal plane
         'phi': half_angle_ODT2,
         'label': 'ODT2'}

# -- setup trap
odt = Trap()
odt.add_laser(**beamODT1)
odt.add_laser(**beamODT2)

# -- analyze
# expected properties
odt.compute_theoretical_properties()
# plot potential
odt.plot_potential(spatial_range=(0.8e-3, 300e-6, 300e-6), figsize=(13,7))

# compute frequency
freq_res = odt.analyze_freq(spatial_range=(60e-6, 20e-6, 20e-6), plot_result=False, only_print_mean=True)
# compute depth
depth_res = odt.analyze_depth(spatial_range=(1e-3, 500e-6, 500e-6), figsize=(13,4))
```

### Final ODT configuration

```python
# -- settings

# factor to diminish total power
eta=0.25/3.5

# parameters
power_ODT1 = 18 * eta
waist_ODT1 = 133e-6
half_angle_ODT1 = 20 * pi / 180  # radians

power_ODT2 = 8 * eta
waist_ODT2 = 63e-6
half_angle_ODT2 = -20 * pi / 180  # radians
 
# gradient coil parameters
grad_coil_current = 5
grad_coil_radius = 11e-2/2
grad_coil_position = -8e-2
grad_coil_plane = 'yz'
grad_coil_turns = 11

# beams
# NB : wavelength is 1550nm by default
beamODT1 = {'waist_value': waist_ODT1,
         'power': power_ODT1,
         'theta': pi / 2,  # i.e. in horizontal plane
         'phi': half_angle_ODT1,
         'label': 'ODT1'}

beamODT2 = {'waist_value': waist_ODT2,
         'power': power_ODT2,
         'theta': pi / 2,  # i.e. in horizontal plane
         'phi': half_angle_ODT2,
         'label': 'ODT2'}

# gradient coil
grad_coil = {'plane': grad_coil_plane,
             'radius': grad_coil_radius,
             'axial_shift': grad_coil_position,
             'n_turns': grad_coil_turns,
             'current': grad_coil_current,
             'label': 'grad. coil'}

# -- setup trap
odt = Trap()
odt.add_laser(**beamODT1)
odt.add_laser(**beamODT2)
odt.add_coil(**grad_coil)

# -- analyze
# expected properties
odt.compute_theoretical_properties()
# plot potential
odt.plot_potential(spatial_range=(0.8e-3, 300e-6, 300e-6), figsize=(13,7))

# compute frequency
freq_res = odt.analyze_freq(spatial_range=(60e-6, 20e-6, 20e-6), plot_result=False, only_print_mean=True)
# compute depth
depth_res = odt.analyze_depth(spatial_range=(1e-3, 200e-6, 200e-6), figsize=(13,4))
```

## Magnetic coils

### Bias fields

```python
from coil import SingleCoil, CoilSet
```

```python

# === Helmholtz configuration - Bias X

# - settings
radius = 35*10**(-2)/2
plane = 'yz'

# - define coils
coil1 = {'plane': plane,
         'current': 1,
         'radius': radius,
         'axial_shift': 18*10**(-2) /2,
         'n_turns': 25
         }

coil2 = {k: v for k, v in coil1.items()}
coil2['axial_shift'] = -coil2['axial_shift']

# - init CoilSet
# it takes a list of dictionnaries
# each dict cointains the settings for one coil
helm_coilX = CoilSet([coil1, coil2], label='HC')

# - grids and computation
# grid
x = np.linspace(-.3, .3, 500)
z = np.linspace(-.3, .3, 500)
y = 0
x, z = np.meshgrid(x, z)

# compute
Bx, By, Bz = helm_coilX.field(x, y, z, unit='G')
Bx0, By0, Bz0 = helm_coilX.field(0, 0, 0, unit='G')
B = np.sqrt(Bx**2 + Bz**2 + By**2)

# print
print('Bias field along z: Bx(0)=%.2f G' % Bx0)
#print('|B|_max = %.2f G' % np.max(B))
#for [ax, Bi] in zip(['x', 'y', 'z'], [Bx, By, Bz]):
#    print('|B%s|_max = %.2f G' % (ax, np.max(Bi)))

# plot

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.streamplot(x, z, Bx, Bz, color='k')
pcm = ax.pcolormesh(x, z, np.sqrt(Bx**2 + Bz**2 + By**2), vmin=0, vmax=2, cmap='Spectral_r')
ax.set_ylabel('Z (m)')
ax.set_xlabel('X (m)')
plt.colorbar(pcm, ax=ax, label='|B| (Gauss)')
plt.show()
```
